```python
import tkinter as tk
import nltk

pos_tags_dict = {
    "CC": "Coordinating conjunction",
    "CD": "Cardinal number",
    "DT": "Determiner",
    "EX": "Existential there",
    "FW": "Foreign word",
    "IN": "Preposition or subordinating conjunction",
    "JJ": "Adjective",
    "JJR": "Adjective, comparative",
    "JJS": "Adjective, superlative",
    "LS": "List item marker",
    "MD": "Modal",
    "NN": "Noun, singular or mass",
    "NNS": "Noun, plural",
    "NNP": "Proper noun, singular",
    "NNPS": "Proper noun, plural",
    "PDT": "Predeterminer",
    "POS": "Possessive ending",
    "PRP": "Personal pronoun",
    "PRP$": "Possessive pronoun",
    "RB": "Adverb",
    "RBR": "Adverb, comparative",
    "RBS": "Adverb, superlative",
    "RP": "Particle",
    "SYM": "Symbol",
    "TO": "to",
    "UH": "Interjection",
    "VB": "Verb, base form",
    "VBD": "Verb, past tense",
    "VBG": "Verb, gerund or present participle",
    "VBN": "Verb, past participle",
    "VBP": "Verb, non-3rd person singular present",
    "VBZ": "Verb, 3rd person singular present",
    "WDT": "Wh-determiner",
    "WP": "Wh-pronoun",
    "WP$": "Possessive wh-pronoun",
    "WRB": "Wh-adverb"
}

def show_full_forms():
    # Create a string with the full form of each POS tagger
    full_forms = "\n".join([f"{tag}: {full_form}" for tag, full_form in pos_tags_dict.items()])

    # Show the full forms in a new window
    full_forms_window = tk.Toplevel(window)
    full_forms_window.title("POS Tag Full Forms")
    full_forms_label = tk.Label(full_forms_window, text=full_forms)
    full_forms_label.pack()

# Define a function to perform POS tagging
def pos_tagging():
    # Get the input sentence from the text box
    sentence = entry.get()
    
    # Tokenize the sentence into words
    words = nltk.word_tokenize(sentence)

    # Perform POS tagging on the words
    pos_tags = nltk.pos_tag(words)
    
    # Format the tagged sentence
    tagged_words = [f"{word}/{tag}" for word, tag in pos_tags]
    tagged_sentence = " ".join(tagged_words)
    tagged_sentence = tagged_sentence.replace(' ,', ',').replace(' .', '.').replace(' ?', '?').replace(' !', '!').replace(' :', ':').replace(' ;', ';')
    
    # Update the output label with the tagged sentence
    output.configure(text=f"Tagged sentence:\n{tagged_sentence}")

# Create the GUI window
window = tk.Tk()
window.title("POS Tagger")
window.geometry("400x300")

# Create the input label and text box
label = tk.Label(window, text="Enter a sentence:")
label.pack()
entry = tk.Entry(window)
entry.pack()

# Create the button to perform POS tagging
button = tk.Button(window, text="Tag", command=pos_tagging)
button.pack()

# Create the output label to display the tagged sentence
output = tk.Label(window, text="")
output.pack()

# Create the full form button
full_form_button = tk.Button(window, text="Full Form", command=show_full_forms)
full_form_button.pack()

# Start the GUI loop
window.mainloop()
```
